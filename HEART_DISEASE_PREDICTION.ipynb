```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#loading the csv data to pandas data file
heart_data = pd.read_csv('/content/Heart_Disease_Prediction.csv')
```

```python
#print first 5 rows of dataset
heart_data.head()
```

```python
#print last 5 rows of dataset
heart_data.tail()
```

```python
#number of rows and columns in dataset
heart_data.shape
```

```python
#getting information about data
heart_data.info()
```

```python
#checking for missing values
heart_data.isnull().sum()
```

```python
#statistical measures abt the data
heart_data.describe()
```

```python
print(heart_data.columns)
```

```python
heart_data['Heart Disease'] = heart_data['Heart Disease'].map({'Presence': 1, 'Absence': 0})
```

```python
heart_data.rename(columns={'Heart Disease': 'target'}, inplace=True)
```

```python
print(heart_data.head())  # Should show 0s and 1s
print(heart_data['target'].value_counts())  # Should show the distribution of 0s and 1s
```

0 --> healthy heart 1 --> defective heart


Splitting the features and target

```python
X = heart_data.drop(columns='target', axis=1)
Y = heart_data['target']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data into training data and test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify = Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

Logistic Regression

```python
model = LogisticRegression()
```

```python
#training the logistic regression model with training data
model.fit(X_train, Y_train)
```

Model Evalutaion

Accuracy Score

```python
#accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on training data: ', training_data_accuracy)
```

```python
#accuracy on test data
X_test_prediction = model.predict(X_train)
test_data_accuracy = accuracy_score(X_test_prediction, Y_train)
```

```python
print('Accuracy on test data: ', test_data_accuracy)
```

Building a Predictive System

```python
input_data = (70,1,4,130,322,0,2,109,0,2.4,2,3,3)

#change input data to a numpy array
input_data_as_numpy_array = np.asarray(input_data)

#reshape the numpy array as we predict for only one instance
input_data_reshaped = input_data_as_numpy_array.reshape(1, -1)

prediction = model.predict(input_data_reshaped)
print(prediction)

if(prediction[0]==0):
  print('The person does not have a Heart Disease')
else:
  print('The person has Heart Disease')
```
